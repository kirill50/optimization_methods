1. Сталепрокатное предприятие выпускает стандартные стальные листы размерами 1500×6000 мм. <br>
Ежемесячный объем заказов на листы нестандартных размеров приведён ниже:<br>
500×600 — 3000, 1000×400 — 2000, 1500×1000 — 1500<br>
Необходимо составить ЦФ для нахождения наименее затратного раскроя листов.<br>
** и составить алгоритм

$$\sum_{i=1}^{m}x_ia_i \rightarrow min$$

**варианты разреза:**<br>
30(500×600) - остаток 0 <br>
10(500×600) + 15(1000×400) - остаток 0<br>
6(1500×1000) - остаток 0 <br>

**Решение:**<br>
134 листа на разрез вариантом №2 = 2010 (1000×400) и 1340 (500×600)<br>
3000 - 1340 = 1660<br>
56 листа на разрез вариантом №1 = 1680 + 1340 = 3020 (500×600)<br>
250 листов на разрез вариантом №3 = 1500 (1500×1000)<br>
<br>
Итого надо 440 листов из которых мы получим 3020 - (500×600), 2010 - (1000×400), 1500 - (1500×1000).<br>

Но это решение "на глаз" методом подгона вручную и он получилось только потому, <br>
что мы имеем ровные для этой задачи размерности. Решить это по нормальному и тем <br>
более предложить алгоритм не могу.

2. Пусть есть компания, занимающаяся доставкой Belivery Club.<br>
Есть N зон, в которых есть спрос на курьеров в каждый час. Внутри дня в разные часы спрос различен.<br>
Каждой зоне i в час h необходимо необходимо $n_{hi}$ курьеров. Для каждой зоны есть штраф из-за <br>
недостаточно 1 курьера - s_i. Так же у каждого курьера есть предпочтения относительно часов и зон работы. <br>
Необходимо распределить M курьеров таким образом, чтобы суммарный штраф был минимальный.<br>
(описать формулу в общем случае, т.е. без каких-либо числовых данных)

Пронумеруем курьеров:
пусть $x_{ih}^k$ = 1, если k-курьер работает в час h в i зоне

Тогда в одной зоне в час h работает $$\sum_{k=1}^{M} x_{ih}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} x_{ih}^k - n_{ih} >0 \\
s_i * (\sum_{k=1}^{M} x_{ih}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 



Штраф за день $$\sum_{h=1}^{24}f_{ih}$$

Штраф по всем зонам в течении дня:


$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}$$

Целевая функция:

$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih} \rightarrow min $$

Ограничения: 
$$x_{ih}^k \in N_{jh} $$

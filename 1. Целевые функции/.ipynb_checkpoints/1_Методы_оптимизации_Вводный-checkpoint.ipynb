## Системы: оптимальные структуры и режимы работы


Большая часть времени и денег — того, что мы ценим больше всего, — уходит на ненужные вещи или идеи. Бывает, что, потратив кучу сил и средств на создание системы с великим множеством входных и выходных данных, окинув взглядом готовый продукт, понимаешь, что мог бы написать в два раза меньше кода, создать на десяток меньше таблиц и индексов, сделать систему более гибкой и изящной.
<br>Любая система призвана выполнять какую-то одну функцию или набор функций. Системы, функционалы которых схожи, объединяются в классы. Системы одного класса, созданные разными разработчиками, выполняют свой набор функций с разной эффективностью. Модификация системы с целью повышения её эффективности называется оптимизацией. Полностью оптимальной считается система, показатели работы которой невозможно улучшить путём внесения изменений в структуру системы или в её настройки. <br>
Эффективность системы могут составлять её полезные характеристики: скорость, пропускная способность, уровень шума, герметичность, плавность хода и так далее. Чем эффективнее система, тем больше экономится денег, времени, сил, средств, нервов потребителя и т. д. Сделать систему лучше, качественнее, ближе к оптимальной помогают методы оптимизации, которые применяются при решении задач оптимизации, возникающих на этапе проектирования или доработки.


## Целевая функция. Критерии оптимальности


Задачи отыскания наибольших или наименьших величин часто возникают в науке,технике и экономике. Чтобы применять математические методы для их решения и анализа, необходимо уметь переходить от содержательной к математической постановке задачи. Для этого нужно определить:
- целевую функцию $F(\vec{x})$: $R^n$ $\rightarrow$ $R$;
- множество допустимых рещений $X \subset R^n $ (допустимое множество) для функции $F(\vec{x})$;
- критерий оптимизации $extr \in \{min, max\}$.

Таким образом,тройка $(F, X, extr)$ задает екстремальную или оптимизационную задачу. Формально математическая постановка выглядит следующим образом:

$F(\vec{x}) \rightarrow extr_{x \in X}$

Целевая функция системы  $F(\vec{x})$ отражает степень эффективности системы и зависит от набора регулируемых либо значимых параметров $\vec{x} = (x_1, x_2, x_3, ..., x_m)$.


Целевая функция (ЦФ) зависит от ряда параметров. В общем случае это входные и выходные величины и параметры регулирования. В качестве параметров регулирования могут выступать величины, которые могут быть на этапе проектирования либо доработки изменены для обеспечения наибольшей эффективности. В основном в качестве ЦФ выбирают зависимость выходных величин от параметров регулирования при условно постоянных входных величинах и внешних воздействиях. Если возможно, желательно выбрать целевую функцию, зависящую от минимально возможного количества параметров для упрощения поиска решения задачи оптимизации. Решением задачи оптимизации будет являться вектор $\vec{x}^* = (x^*_1, x^*_2, x^*_3, ..., x^*_m)$ и значение функции $F(\vec{x}^*)$. 

Возьмем частный случай ЦФ, зависящей от одного параметра, например $F(x)=x^2+1$. Пусть параметр $x \in [-1;1]$, а в качестве критерия оптимальности выбран минимум функции $F(x)$. Тогда очевидно, что $x^*=0$, а $F(x^*) = 1$. Если в качестве критерия оптимальности выбран максимум $F(x)$, то решения будет два: $x_1^* = -1, x_2^* = 1$. Промежуток, в котором может регулироваться $x$, называется допустимым множеством или областью допустимых решений (ОДР).

```python
import numpy as np              
import matplotlib.pyplot as plt 

x = np.arange(-1.0, 1.1, 0.1)
F = x ** 2 + 1                # выражение ЦФ
fig, ax = plt.subplots()      

ax.plot(x, F)   
ax.grid()       
 
lgnd = ax.legend(['F'], loc='upper center', shadow=True) 
lgnd.get_frame().set_facecolor('#ffb19a')                
 
plt.show()                    # вывод графика
```

## Гладкие ЦФ. Комбинаторные ЦФ

Гладкой называется функция, имеющая непрерывную производную на всей ОДР. Если функция непрерывно дифференцируема, то задача поиска экстремумов (прим.: максимумов и минимумов) сводится к нахождению нулей частных производных по всем переменным.<br>

Для примера возьмем некую ЦФ с критерием максимальности, определенную на участке $[0,5; 2]$ :<br>
<center>$$F(x)=\frac{2\sqrt[3]{x}-x^x+1,5x+1}{3}$$</center><br>
Очевидно, что функция является дифференцируемой на всей ОДР. Её производная будет равна:<br>
<center>$$\frac{dF(x)}{dx}=\frac{2}{9\sqrt[3]{x^2}}-\frac{x^x(\ln{x+1})}{3}+0,5$$</center><br>
C помощью pyplot построим графики обеих функций на ОДР.

```python
import numpy as np              
import matplotlib.pyplot as plt 
import math

x = np.arange(0.5, 2.0, 0.01)                                          # ряд от -0,5 до 2,0 с шагом 0,01
F = (2*x ** 1/3 - x**x + 1.5*x + 1)/3                                  # выражение ЦФ
dF = 2/9*x ** (-2/3) - x**x*(np.log(x)+1)/3 + 0.5                      # выражение производной ЦФ
fig, ax = plt.subplots()                                               # прорисовка осей

ax.plot(x, F)                                                          # прорисовка графика ЦФ
ax.plot(x, dF)                                                         # прорисовка графика производной
plt.axvline(x=1.39)
ax.grid()                                                              # прорисовка сетки
 
lgnd = ax.legend(['ЦФ', 'Производная'], loc='lower left', shadow=True) # легенда
lgnd.get_frame().set_facecolor('#ffb19a')                              # рамка легенды
 
plt.show()                                                             # вывод графика

for i in range(85, 95):
    print("%.2f %.3f %.3f\n"%(x[i], F[i], dF[i]))
```

По графикам и по табуляции видно, что производная меняет знак на промежутке $(1,37; 1,38)$, а функция начинает убывать на $(1,40; 1,41)$. Расхождение в результатах можно списать на точность вычисления логарифма, а решением задачи оптимизации по критерию максимума считать средним установленных величин (около 1,39).

В этом уроке приведён пример гладкой функции без привязки к реальному миру. В реальности правильный и точный подбор целевой функции обеспечивает верное решение задачи оптимизации вне зависимости от выбранного метода решения.

Комбинаторные ЦФ используются для решения задач комбинаторной оптимизации. Ярким примером такой задачи является задача коммивояжёра. 
Имеется $n$ городов. Выезжая из одного, комиваяжер должен объехать все и вернуться в исходных город. В каждый город можно заезжать только один раз. Требуется найти минимальный по стоимости или протяженности замкнутый маршрут.

Как правило, ЦФ для этой задачи служит длина гамильтоновой цикла (цикл, который проходит через каждую вершину в графе, в нашем случае город). Оптимизация происходит путём частичного или полного изменения маршрута по одному или нескольким правилам. 

## Построение ЦФ для реальной системы

Допустим, необходимо проложить тепловую магистраль диаметром 20 см и протяженностью 100 км. Температура воды на выходе из ТЭЦ — $T_0 = 90^{\circ}C$. Эксплуатационная температура — $T_{эксп} = -20^{\circ}C$. Чем выше давление в магистрали, тем больше расчётная толщина стенки трубы:<br>
<center>$$S = \frac{PD_a}{2\sigma+P}$$</center>
Здесь $S$ — толщина стенки, $D_a$ — диаметр трубы (наружный), $P$ — давление.<br>
Чем выше давление в магистрали, тем выше скорость течения воды, тем меньше теплоотдача:
<center>$$T_{100}=T_0-\frac{T_0-T_{эксп}}{PS}$$</center>
Один слой теплоизоляционного материала на километр магистрали стоит 1000 рублей и позволяет сохранить в контрольной точке 
20% от возможной теплопотери. Стоимость одного квадратного миллиметра на километр трубы стоит 100 рублей.
<br>Чем больше толщина стенки, тем меньше пропускная способность магистрали. Допустимая теплопотеря — $10^{\circ}C$
<br>Необходимо спроектировать тепловую магистраль с наименьшими материалозатратами и с минимальной теплопотерей.

```python
import numpy as np              
import matplotlib.pyplot as plt 
import math

P = np.arange(9, 11, 0.01)                           # ряд от 9 до 11 с шагом 0,1
S = P*200/(2*180+P)                                  # выражение для толщины стенки
T100 = 90 - 70/(P*S)                            # выражение для температуры в контрольной точке для магистрали без изоляции
dT = (90 - T100)*1000                                # выражения для теплопотери для магистрали без изоляции, масштаб 1000

costp = 0.005*100*100*(400 - (20 - S)*(20 - S))*3.14  # стоимость зависит от площади сечения, умножена на к-т масштаба 0,005
dT2 = dT*0.8                                         # намотка одного слоя изоляции снижает теплоотдачу на 20%, масштаб 1000
costi = 1000*100*0.005                               # cтоимость намотки одного слоя изоляции умножена на к-т масштаба 0,005
costf = costp + costi

TF1 = (np.abs(dT) + costp / 100) * 10     
TF2 = (np.abs(dT2) + costf / 100) * 10   
TF = np.minimum(TF1, TF2)                            # целевая функция
# рассматриваются точки с максимально допустимой теплопотерей;
# в этих точках оценивается стоимость решений
# постоянные и множители взяты с учётом масштабов

fig, ax = plt.subplots()  
fig.set_size_inches(9,6)
# прорисовка осей

ax.plot(P, costp)                                                                                # прорисовка графика стоимости материалов
ax.plot(P, costf)                                                                                # прорисовка графика стоимости материалов c изоляцией
ax.plot(P, dT)                                                                                   # прорисовка графика теплопотерь
ax.plot(P, dT2)                                                                                  # прорисовка графика теплопотерь с изоляцией
ax.plot(P, TF)                                                                                   # прорисовка графика целевой функции

plt.axhline(y=10000)                                                                             # допустимая теплопотеря 10 градусов Цельсия
plt.axvline(x=10.545)                                                                            # второе оптимальное давление
plt.axvline(x=9.415)                                                                             # первое оптимальное давление
ax.grid()                                                                                        # прорисовка сетки
 
lgnd = ax.legend(['стоимость', 'стоимость с изоляцией', 'теплопотеря',\
                  'теплопотеря с изоляцией','целевая функция'], loc='upper center', shadow=True) # легенда
lgnd.get_frame().set_facecolor('#ffb19a')                                                        # рамка легенды
 
plt.show()                                                                                       # вывод графика
```

Здесь сформулирована задача с огромным количеством допущений и приближений. Перед вводом каждого приближения необходимо оценивать его возможное влияние на целевую функцию. К задаче оптимизации может быть составлена целевая функция, выбран критерий оптимизаций и применён любой метод для её решения, но если целевая функция выбрана с ошибкой, дальнейшие операции вне зависимости от их правильности смысла иметь не будут.
